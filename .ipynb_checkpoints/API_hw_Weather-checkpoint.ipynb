```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

from config import api_key

from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

<p>This code uses the numpy random method to generate numbers in the range of the globes system of latitude and logitude. It then zips an array of these latitudes and longitudes.</p>

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
```

<p>Having set up a list of cities, the below script will feed the sliced values in the lat_lngs array through the citypy library to populate the list with real city names. I also included the len printout of the cities list to compare to results below.</p>

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

<p>In the below cell, I set up an f-string of the query url having defined the openweathermap url, units as metric and the api_key as a line in a .py file. I also defined five lists of weather attributes and latitude. With a for loop, the script will iterate through the list of cities, "get" weather data from openweathermap, convert the data to .json and then append the five attributes to their respective lists.
</p>
<p>An issue that I had was that a portion of cities returned from citypy do not correspond exactly to cities on the openweathermap. I first included a .remove method at the end of the loop to remove the city name from the list but this line also removed one value from each of the attribute lists. While debugging I decided to leave the city name in the cities list and append the five attribute lists with a string of nonsense.</p>
<p>I will deal with this non-numerical data below. I included a series of len functions to demonstrate that the code words by show that each lists were of equal length. This numbers, now equal can be compared to the number above to see that the number in the cities lists iterated correctly.</p>

```python
#setting up variables for query_url f-string
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

#defining lists of weather attributes
lat = []
temp = []
hum = []
wind = []
cloud = []

#for loop to iterate through cities list, "get" weather data, convert data to .json and append five 
#attributes to their respective lists. In the case of a citypy city which is not in the openweathermap
#the else portion appends nonsense to the five lists
for city in cities:
        query_url = f"{url}appid={api_key}&q={city}&units={units}"
        weather_request = requests.get(query_url)
        
        if weather_request:          
            weather_json = weather_request.json()
            lat.append(weather_json['coord']['lat'])
            temp.append(weather_json['main']['temp'])
            hum.append(weather_json['main']['humidity'])
            wind.append(weather_json['wind']['speed'])
            cloud.append(weather_json['clouds']['all']);
        else:
            (weather_json['coord']['lat']) = "xxx"
            (weather_json['main']['temp']) = "xxx"
            (weather_json['main']['humidity']) = "xxx"
            (weather_json['wind']['speed']) = "xxx"
            (weather_json['clouds']['all']) = "xxx"
            lat.append(weather_json['coord']['lat'])
            temp.append(weather_json['main']['temp'])
            hum.append(weather_json['main']['humidity'])
            wind.append(weather_json['wind']['speed'])
            cloud.append(weather_json['clouds']['all']);

len(cities)
len(lat)
len(temp)
len(hum)
len(wind)
len(cloud)
```

```python
#creating a dictionary from the above cities and attributes lists to form the basis of a pandas dataframe
weather_dict = {
    "City": cities,
    "Latitude": lat,
    "Temperature": temp,
    "Humidity": hum,
    "WindSpeed":wind,
    "Cloudiness": cloud
}
# weather_dict
```

```python
#creates the dataframe to use for scatter plots. 
#eliminates entire lines with no weather data
#compare number of lines in the df to the original len of cities list. The difference being
#the number of city names in conflict between citypy and openweathermap in this iteration
weather_df = pd.DataFrame(weather_dict)
weather_df = weather_df[weather_df.Latitude != "xxx"]
weather_df.head(2)
```

<h3>Weather Plots: Temperature, Humdity, Wind Speed and Cloudiness against global latitude</h3>

```python
weather_df.plot(x="Latitude",y="Temperature",linestyle="",marker=".",figsize=(20,10))

plt.grid(color="lightgreen")
plt.title("Temperature across the globe")
plt.xlabel("Latitude")
plt.ylabel("Temperature (degC)")
plt.show();
```

```python
weather_df.plot(x="Latitude",y="Humidity",linestyle="",marker=".",figsize=(20,10))

plt.grid(color="lightblue")
plt.title("Humdity across the globe")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.show();
```

```python
weather_df.plot(x="Latitude",y="WindSpeed",linestyle="",marker=".",figsize=(20,10))

plt.grid(color="pink")
plt.title("Wind Speed across the globe")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (kph))")
plt.show();
```

```python
weather_df.plot(x="Latitude",y="Cloudiness",linestyle="",marker=".",figsize=(20,10))

plt.grid(color="grey")
plt.title("Cloudiness across the globe")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.show();
```
